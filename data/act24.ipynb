```python
import pandas as pd
df = pd.read_csv('UKStatistics.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df
```

```python
def my_hist(column, title, xlab, ylab=' Count', color='green',
            alpha=0.6, bins=10):
    plt.figure(figsize=(14,10))
    plt.hist(column, color=color, alpha=alpha)
    plt.title(title, fontsize=15)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.savefig(title, dpi=300)
    plt.show
```

```python
my_hist(df["Actual Pay Floor (£)"],"Actual Pay Floor","amount")
```

```python
plt.figure(figsize=(16,10))
my_title='Actual Pay Floor Scatterplot'
plt.title(my_title, size=15)
```

```python
sns.scatterplot(x=df['Salary Cost of Reports (£)'], 
                y=df['Actual Pay Floor (£)'],
                sizes=(20, 400))
plt.savefig(my_title, dpi=225)
plt.show()
```

```python
 plt.figure(figsize=(14, 10))
 title='Actual Pay Floor'
 plt.title(title, size=15)
 sns.boxplot(x = df['Salary Cost of Reports (£)'])
 plt.savefig(title, dpi=300)
 plt.show()
```
