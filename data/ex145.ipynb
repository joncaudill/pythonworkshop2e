```python
import pandas as pd
df = pd.read_csv('HousingData.csv')

df.corr()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
corr = df.corr()
plt.figure(figsize=(14,10))
sns.heatmap(corr, xticklabels=corr.columns.values,
            yticklabels=corr.columns.values,
            cmap="Reds",
            linewidths=1.25)
plt.show()
```

```python
x = df['RM']
y = df['MEDV']
plt.figure(figsize=(14, 10))
sns.regplot(x=x,y=y)
plt.show()
```

```python
import statsmodels.api as sm
X = sm.add_constant(x)
model = sm.OLS(y, X)
est = model.fit()
print(est.summary())
```
