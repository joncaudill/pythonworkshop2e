```python
import pandas as pd
df = pd.read_csv('HousingData.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
title = 'Median Boston Housing Prices'
```

```python
plt.figure(figsize=(14,8))
```

```python
plt.hist(df['MEDV'], color='green', alpha=0.6)
```

```python
plt.title(title, fontsize=15)
plt.xlabel('1980 Median Value in Thousands')
plt.ylabel('Count')
plt.hist(df['MEDV'], color='green', alpha=0.6)
plt.savefig(title, dpi=300)
plt.show()
```

```python
def my_hist(column, title, xlab, ylab=' Count', color= 'green', alpha=0.6, bins=10, range=None):
    plt.figure(figsize=(14,8))
    plt.hist(column, color=color, alpha=alpha, bins=bins, range=range)
    plt.title(title, fontsize=15)
    plt.xlabel(xlab)
    plt.ylabel(ylab)
    plt.savefig(title, dpi=300)
    plt.show()
```

```python
my_hist(df['RM'], 
        'Average Number of Rooms in Boston Households', 
        'Average Number of Rooms', 
        color='skyblue',
        bins=6,
        range=(3,9)
       )
```
